```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
athlete = pd.read_csv('athlete_events.csv')
region = pd.read_csv('noc_regions.csv') 
```

```python
athlete.head()
```

```python
region.head()
```

```python
# join DB
```

```python
athlete_df = athlete.merge(region, how = 'left', on = 'NOC')
athlete_df.head()
```

```python
 athlete_df.shape
```

```python
# colunm name consistent
athlete_df.rename(columns= {'region': 'Region', 'notes': 'Notes'}, inplace= True)
```

```python
# Check null values
nan_value = athlete_df.isna()
nan_columns = nan_value.any()
nan_columns
```

```python
# print the columns names containing null values or missing values in the form of a list
athlete_df.columns[athlete_df.isnull().any()]
```

```python
# Israel details about participating
athlete_df.query('Team == "Israel"')
```

```python
# top 10 countries participating
top_10 = athlete_df.Team.value_counts().sort_values(ascending=False).head(10)
top_10
```

```python
# plot for top 10 countries

plt.figure(figsize= (12, 6))
plt.xticks(rotation=20)
plt.title('Overall Participation by Country')
sns.barplot(x= top_10.index, y= top_10, palette= 'Set2')
```

```python
# Age Distribution of the participents
plt.figure(figsize=(12, 6))
plt.title('Age Distribution of the athlets')
plt.xlabel('Age')
plt.ylabel('Nember of participents')
plt.hist(athlete_df.Age, bins = np.arange(10,80,2),color='orange', edgecolor = 'white');
```

```python
# Male and Female participants
gender_counter = athlete_df.Sex.value_counts()
gender_counter
```

```python
# Pie plot for male and female athletes

plt.figure(figsize=(12, 6))
plt.title('Gender Distribution')
plt.pie(gender_counter, labels=gender_counter.index, autopct='%2.2f%%', startangle=150, shadow=True);
```

```python
# Total medals

athlete_df.Medal.value_counts()
```

```python
# Total number of female in each olympics

female_counter = athlete_df[(athlete_df.Sex == 'F') & (athlete_df.Season == 'Summer')][['Sex', 'Year']]
female_counter = female_counter.groupby('Year').count().reset_index()
female_counter.tail()
```

```python
female_in_olympics = athlete_df[(athlete_df.Sex == 'F') & (athlete_df.Season == 'Summer')]
```

```python
sns.set(style = 'darkgrid')
plt.figure(figsize=(20, 10))
sns.countplot(x='Year', data= female_in_olympics, palette='Spectral')
plt.title('Woman Athletes over the years')
```

```python
part = female_in_olympics.groupby('Year')['Sex'].value_counts()
plt.figure(figsize=(20,10))
part.loc[:,'F'].plot()
plt.title('Plot of Female Athletes over time')
```

```python
# Gold medal athletes

gold = athlete_df[(athlete_df.Medal == 'Gold')]
gold = gold[np.isfinite(gold['Age'])]
```

```python
gold['ID'][gold['Age'] > 60].count()
```

```python
sporting_event = gold['Sport'][gold['Age']>60]
sporting_event
```

```python
plt.figure(figsize=(10,5))
plt.tight_layout()
sns.countplot(sporting_event)
plt.title('Gold Medals for athletes over 60')
```

```python
# Gold from each country
total_gold = gold.Region.value_counts().reset_index(name='Medal').head(5)
total_gold
```

```python

g= sns.catplot(x='index', y='Medal', data=total_gold, height=5, kind='bar', palette='rocket')
g.despine(left=True)
g.set_xlabels('Top 5 countries')
g.set_ylabels('Number of Medals')
plt.title('Gold Medals per country')
```

```python
# Rio olympic

max_year = athlete_df.Year.max()
max_year

teams = athlete_df[(athlete_df.Year == max_year) & (athlete_df.Medal == 'Gold')].Team
teams.value_counts().head(10)
```

```python
sns.barplot(x=teams.value_counts().head(20), y=teams.value_counts().head(20).index)

plt.ylabel(None)
plt.xlabel('Countrywise Medals for the year 2016');
```

```python
real_medals = athlete_df[(athlete_df['Height'].notnull()) & (athlete_df['Weight'].notnull())];
```

```python
# 'Comparison between Israli men and women in the olympics games'
israel_athletes = athlete_df[(athlete_df['NOC']=='ISR')].Sex.value_counts()
israel_athletes
```

```python
plt.figure(figsize=(20, 10))
plt.title('Comparison between Israli men and women in the olympics games')
myexplode = [0.2, 0]
labels=israel_athletes.index

plt.pie(israel_athletes, labels=labels, explode = myexplode, autopct='%2.1f%%', startangle=100,
        shadow=True, colors = 'mycolors');
plt.legend(title = 'gender')
```
